```julia
using JLD2
using StatsPlots, Plots, Plots.PlotMeasures
```

```julia
f = jldopen("/Users/joshua/Documents/QuantumInspiredML/Sampling/benchmarking/nslvn/fourier_scores.jld2", "r");
scores = f["scores"]
```

```julia
keys(scores)
```

```julia
mse_scores = Dict{String, Vector{Float64}}()
for (key, value) in scores
    lower_key = lowercase(key)
    mse_values = []
    for i in 1:length(value)
        mse = value[i][2][:MSE]
        push!(mse_values, mse)
    end
    mse_scores[lower_key] = mse_values
end
```

```julia
mse_scores
```

```julia
boxplot(mse_scores["fourier_d2_chi10"], label="d=2")
boxplot!(mse_scores["fourier_d3_chi10"], label="d=3")
boxplot!(mse_scores["fourier_d4_chi10"], label="d=4")
boxplot!(mse_scores["fourier_d5_chi10"], label="d=5")
boxplot!(mse_scores["fourier_d6_chi10"], label="d=6")
```
