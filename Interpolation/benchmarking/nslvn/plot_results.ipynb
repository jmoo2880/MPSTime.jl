```julia
using JLD2
using StatsPlots, Plots, Plots.PlotMeasures
using StatsBase
```

```julia
f = jldopen("/Users/joshua/QuantumMay/QuantumInspiredML/Interpolation/benchmarking/ecg200/fourier/fourier_70_site_results.jld2", "r");
scores = f["fourier_70_site_results"]
```

```julia
keys(scores)
```

```julia
mse_scores = Dict{String, Vector{Float64}}()
for (key, value) in scores
    lower_key = lowercase(key)
    mse_values = [x[:MSE] for x in value]
    mse_scores[lower_key] = mse_values
end
```
